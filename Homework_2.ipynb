```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Assuming `df` is your DataFrame loaded with the appropriate data

# Target and feature variables
target_variable = 'Final Price'  # Ensure this matches your column name
feature_columns = ['RAM', 'Storage', 'Screen']  # Replace with your actual feature columns

# Store RMSE scores
rmse_scores = []

# Try different seed values
for seed in range(10):
    # Step 1: Split the data
    train_df, temp_df = train_test_split(df, test_size=0.4, random_state=seed)  # 60% training data
    val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=seed)  # 20% validation and 20% testing data

    # Step 2: Fill missing values with 0
    train_df.fillna(0, inplace=True)
    val_df.fillna(0, inplace=True)
    
    # Step 3: Prepare features (X) and target (y)
    X_train = train_df[feature_columns]
    y_train = train_df[target_variable]
    X_val = val_df[feature_columns]
    y_val = val_df[target_variable]

    # Step 4: Train the linear regression model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Step 5: Predict and calculate RMSE
    y_pred = model.predict(X_val)
    rmse = mean_squared_error(y_val, y_pred, squared=False)
    rmse_scores.append(rmse)

# Step 6: Calculate the standard deviation of RMSE scores
std_rmse = np.round(np.std(rmse_scores), 3)

# Output the standard deviation
print("Standard Deviation of RMSE Scores:", std_rmse)
```
